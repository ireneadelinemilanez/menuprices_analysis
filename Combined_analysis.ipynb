```python
## import library
import pandas as pd
```

```python
## pull in data
df = pd.read_csv("combined1.csv")
df
```

```python
## delete empty rows
df = df.dropna()
df
```

```python
### Lambda expression, .iat = index at, holds a row value as an index starting point
```

```python
### Hold CPI, median income, median house

cpi_start = df["cpi_away_food"].iat[0]
cpi_start
```

```python
pers_start = df["med_inc_pers"].iat[0]
pers_start
```

```python
house_start = df["med_inc_house"].iat[0]
house_start
```

```python
## create a new column that holds the zero index for columns

df["cpi_zero_index"] = df["cpi_away_food"].apply(lambda x:round((x-cpi_start)/cpi_start*100, 2))
df
```

```python
df["pers_zero_index"] = df["med_inc_pers"].apply(lambda x:round((x-pers_start)/pers_start*100, 2))
df
```

```python
df["house_zero_index"] = df["med_inc_house"].apply(lambda x:round((x-house_start)/house_start*100, 2))
df
```
