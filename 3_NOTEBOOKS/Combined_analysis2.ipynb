```python
## import library
import pandas as pd
```

```python
df = pd.read_csv("combined2.csv")
df
```

```python
cpi_start = df["cpi_away_food"].iat[0]
cpi_start
```

```python
house_start = df["med_inc_house"].iat[0]
house_start
```

```python
chipotle_start = df["avg_chipotle"].iat[0]
chipotle_start
```

```python
df["cpi_zero_index"] = df["cpi_away_food"].apply(lambda x:round((x-cpi_start)/cpi_start*100, 2))
df
```

```python
df["house_zero_index"] = df["med_inc_house"].apply(lambda x:round((x-house_start)/house_start*100, 2))
df
```

```python
df["chipotle_zero_index"] = df["avg_chipotle"].apply(lambda x:round((x-chipotle_start)/chipotle_start*100, 2))
df
```

```python
df.to_csv('combined_analysis.csv', index=False)
```
